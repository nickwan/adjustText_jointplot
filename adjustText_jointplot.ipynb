```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import string
import random
from adjustText import adjust_text
```

```python
labels = pd.DataFrame([random.choice(string.ascii_letters) for x in range(50)], columns=['label'])
df = pd.DataFrame(np.random.rand(50,2), columns=['x', 'y'])
df = pd.concat([df, labels], axis=1)
```

```python
def apply_text(df):
    g.ax_joint.text(df['x'], df['y'], df['label'])

g = sns.jointplot('x', 'y', df)
df.apply(apply_text, axis=1)
plt.show()
```

```python
h = sns.jointplot('x', 'y', df)

texts = []
for x,y,s in zip(df['x'], df['y'], df['label']):
    texts.append(h.ax_joint.text(x,y,s))
    
adjust_text(texts)

plt.show()
```
